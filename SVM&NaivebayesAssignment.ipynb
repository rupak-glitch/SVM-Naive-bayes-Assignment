# 1.  What is a Support Vector Machine (SVM)?
   - A Support Vector Machine (SVM) is a supervised machine learning algorithm that can be used for both classification and regression tasks. Its fundamental idea is to find the best possible decision boundary that separates different classes of data. This boundary, called a hyperplane, is chosen in such a way that the margin between the boundary and the closest data points is maximized. A larger margin generally leads to better generalization on unseen data, which is why SVM focuses on maximizing this margin.

# 2. What is the difference between Hard Margin and Soft Margin SVM ?
   - Hard Margin SVM assumes that the data is perfectly linearly separable, meaning no point lies on the wrong side of the boundary and no misclassification is allowed. Since real-world data usually contains noise and overlapping classes, Hard Margin SVM is rarely practical.
   -  Soft Margin SVM, on the other hand, allows some degree of misclassification. It introduces a penalty for points that lie within the margin or on the wrong side of the hyperplane. This creates a balance between having a wide margin and minimizing classification errors, making Soft Margin SVM much more useful for real-world

# 3. What is the mathematical intuition behind SVM ?
 - The mathematical intuition behind SVM is based on an optimization problem. If the separating hyperplane is expressed as w⋅x+b=0, then SVM tries to minimize	1/2(​∥w∥**2) subject to the condition that every data point is correctly classified and lies on the correct side of the margin. For soft margin SVM, slack variables are introduced to allow misclassified or marginal points. In simpler terms, the goal is to keep the model as simple as possible (by minimizing the length of the weight vector) while still classifying the data correctly.

# 4.  What is the role of Lagrange Multipliers in SVM ?
   - Lagrange Multipliers play a crucial role in solving this optimization efficiently. They help convert the constrained optimization problem into what is known as the dual problem. This dual form depends only on dot products between data points rather than the data points themselves. The use of Lagrange multipliers also makes the kernel trick possible, allowing SVMs to deal with non-linear data. Interestingly, the values of these multipliers determine which data points become support vectors.



# 5.  What are Support Vectors in SVM ?
  - Support vectors are the data points that lie closest to the decision boundary. They are extremely important because the position of the hyperplane depends entirely on them. Even if you remove all other data points, the boundary will remain the same as long as the support vectors are unchanged. These points essentially "support" the final decision boundary, which is why the algorithm is named Support Vector Machine.

# 6.  What is a Support Vector Classifier (SVC) ?
  - When SVM is used specifically for classification, it is called a Support Vector Classifier (SVC). The objective is to find the best separating hyperplane between different classes.

# 7.  What is a Support Vector Regressor (SVR) ?
  - When SVM is used specifically for regression, it is known as Support Vector Regressor (SVR). Instead of trying to classify points, SVR tries to fit a function within a margin known as the ε-tube. Points that fall outside this tube become support vectors, and the model attempts to keep the regression function as smooth and flat as possible while allowing small deviations within the ε margin.

# 8.  What is the Kernel Trick in SVM ?
  - Non-linear datasets cannot be separated by a simple straight line or hyperplane. To handle such cases, SVM uses something called the kernel trick. The kernel trick allows the algorithm to implicitly map the data into a higher-dimensional space without actually computing that transformation manually. In this new space, the data may become linearly separable. Kernels compute similarities between data points, allowing SVM to learn complex, curved decision boundaries. Common kernels include the linear kernel, the polynomial kernel, and the radial basis function (RBF) kernel


# 9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel .
  - The linear kernel is used when the data is approximately linearly separable or when the number of features is very large, such as in text classification.
  - The polynomial kernel captures more complex relationships by fitting polynomial-shaped boundaries.
  - The RBF kernel is the most flexible and widely used because it can model highly non-linear decision boundaries by considering the distance between data points in a smooth, Gaussian manner.

# 10.  What is the effect of the C parameter in SVM ?
- The parameter C in SVM controls the balance between having a wide margin and minimizing classification errors. A large value of C forces the model to classify every training example correctly, which results in a smaller margin and a higher risk of overfitting. A smaller value of C allows more misclassified points but results in a larger margin, leading to a model that generalizes better but may underfit if C is too small. Thus, the parameter C serves as a regularization factor that manages the trade-off between flexibility and simplicity.

# 11. What is the role of the Gamma parameter in RBF Kernel SVM ?
  - The gamma parameter in an RBF Kernel SVM controls how far the influence of a single training example reaches. In simple terms, gamma determines the curvature and complexity of the decision boundary. A low value of gamma means that each point has a broad area of influence, which results in a smoother and more generalized decision boundary because the model considers points that are far apart as still somewhat related. This often prevents overfitting but can cause underfitting if gamma is too low. On the other hand, a high value of gamma means that each point has a very narrow influence, making the model focus only on points very close to one another. This can create a highly flexible and detailed decision boundary that captures even small variations in the data. However, if gamma is too high, the model becomes overly sensitive to individual points, leading to overfitting. In summary, gamma controls the sensitivity of the RBF kernel to distances between data points, and choosing the right gamma is essential for achieving a good balance between model complexity and generalization.

# 12.  What is the Naïve Bayes classifier, and why is it called "Naïve" ?
- Naïve Bayes is a probabilistic classifier that uses Bayes’ Theorem to predict the class of a data point by calculating the probability of each class given its features. It is widely used in text classification, spam detection, and sentiment analysis because it is simple, fast, and works well even with high-dimensional data. 
- It is called “Naïve” because it assumes that all features are completely independent of each other, an assumption that is rarely true in real data. Despite this unrealistic assumption, the model still performs surprisingly well in many practical applications.

# 13.  What is Bayes’ Theorem ?
Bayes’ Theorem is a fundamental rule in probability that helps us calculate the probability of an event based on prior knowledge of related conditions. It states that the probability of event A occurring given that event B has occurred is equal to the probability of B given A multiplied by the probability of A, divided by the probability of B. In mathematical form, it is written as 
P(A∣B)=(P(B∣A)⋅P(A))/p(B). Bayes’ Theorem is widely used in machine learning and statistics because it allows us to update our beliefs or predictions as new information becomes available.

# 14. Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes .
  - Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes are three variants of the Naïve Bayes algorithm, each designed for different types of data.
  - Gaussian Naïve Bayes is used when the features are continuous and follow a normal distribution, making it suitable for datasets like sensor readings or any numerical attributes. 
   - Multinomial Naïve Bayes is used for count-based features, especially in text classification, where the data represents word frequencies or term counts.
  -  Bernoulli Naïve Bayes, on the other hand, is used when features are binary, meaning they take values like 0 or 1 to indicate the presence or absence of a word or attribute.

# 15. When should you use Gaussian Naïve Bayes over other variants ?
- Gaussian Naïve Bayes should be used when your features are continuous numerical values and follow, or approximately follow, a normal (Gaussian) distribution. It is especially useful in datasets such as sensor measurements, physical quantities, exam scores, or any situation where values vary smoothly rather than appearing as counts or binary indicators. In these cases, Gaussian Naïve Bayes models the data more naturally and performs better than Multinomial or Bernoulli variants, which are meant for discrete and binary data respectively.

# 16.  What are the key assumptions made by Naïve Bayes ?
  - Naïve Bayes makes two main assumptions about the data in order to simplify the probability calculations. The first assumption is that all features are conditionally independent of one another given the class label, meaning each feature contributes to the prediction independently. The second assumption is that each feature follows a specific probability distribution depending on the variant used, such as a Gaussian distribution for continuous data or a multinomial or Bernoulli distribution for text-based features. These assumptions are rarely true in real-world datasets, but they make the model fast, simple, and surprisingly effective in many practical situations.
 

# 17. What are the advantages and disadvantages of Naïve Bayes ?
- Naïve Bayes has several advantages that make it widely used in machine learning. It is extremely fast, easy to implement, and works well even with very large and high-dimensional datasets, especially in text classification tasks like spam detection or sentiment analysis. It also performs surprisingly well despite its strong independence assumption and requires only small amounts of training data. However, 
- the main disadvantage is that the assumption of feature independence is often unrealistic, which can reduce accuracy when features are strongly correlated. Additionally, Naïve Bayes struggles when a feature value appears in the test data but was never seen in the training data, a problem known as zero probability unless smoothing is applied. Despite these limitations, it remains a simple and effective baseline model.

# 18.  Why is Naïve Bayes a good choice for text classification ?
 - Naïve Bayes is a good choice for text classification because it handles high-dimensional data extremely well, which is important since text datasets often contain thousands of words as features. It is fast to train, requires little data, and works effectively with word frequencies and binary word indicators, which align naturally with the assumptions of Multinomial and Bernoulli Naïve Bayes. Even though the independence assumption is not perfectly true for text, Naïve Bayes still produces strong and reliable results, making it one of the most popular algorithms for tasks like spam filtering, sentiment analysis, and document categorization.

# 19.   Compare SVM and Naïve Bayes for classification tasks 
 - Support Vector Machines (SVM) and Naïve Bayes are both popular classification algorithms, but they differ in their approach and strengths. SVM is a discriminative model that tries to find the optimal hyperplane separating different classes by maximizing the margin between them, making it effective for complex and high-dimensional data, especially when classes are not linearly separable and kernels can be applied. It can achieve high accuracy but often requires more computational resources and careful parameter tuning.
 -  Naïve Bayes, on the other hand, is a probabilistic generative model that predicts class membership based on feature probabilities and assumes that features are conditionally independent given the class. It is simple, fast, and works well for high-dimensional data such as text, even with relatively small training datasets, but its accuracy can suffer if the independence assumption is strongly violated. In general, SVM tends to perform better on numeric or structured datasets with clear margins between classes, while Naïve Bayes is especially effective for text classification and other problems with categorical or count-based features.

# 20.  How does Laplace Smoothing help in Naïve Bayes?
- Laplace Smoothing is used in Naïve Bayes to handle the problem of zero probability, which occurs when a feature value appears in the test data but was never observed in the training data. Without smoothing, this zero probability would make the entire probability of a class become zero, leading to incorrect predictions. Laplace Smoothing solves this by adding a small constant (usually 1) to all feature counts, ensuring that every possible feature value has a non-zero probability. This adjustment allows the model to make more robust predictions, especially in cases where the dataset is small or some feature combinations are rare, without significantly affecting the probabilities of other features.

# 22.  Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then compare their accuracies

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
data = load_wine()
X = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=1)
linear_svm = SVC(kernel='linear')
linear_svm.fit(X_train, y_train)
y_pred_linear = linear_svm.predict(X_test)
linear_accuracy = accuracy_score(y_test, y_pred_linear)
rbf_svm = SVC(kernel='rbf')
rbf_svm.fit(X_train, y_train)
y_pred_rbf = rbf_svm.predict(X_test)
rbf_accuracy = accuracy_score(y_test, y_pred_rbf)
print("Accuracy with Linear Kernel:", linear_accuracy)
print("Accuracy with RBF Kernel:", rbf_accuracy)
```

# 24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

X, y = datasets.make_circles(n_samples=300, factor=0.3, noise=0.05, random_state=0)
model = SVC(kernel='poly', degree=3, C=1)
model.fit(X, y)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))


Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)


plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
plt.title("SVM with Polynomial Kernel (degree = 3)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

# 26.Write a Python program to train a Multinomial Naïve Bayes classifier for text classification using the 20 Newsgroups dataset.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


X = [
    "Space is vast and beautiful",
    "Religion brings people together",
    "Graphics cards are used in computers",
    "Atheism questions religious beliefs",
    "NASA launches rockets to space",
    "Computer graphics are amazing"
]
y = [1, 0, 2, 0, 1, 2]  # 0=Religion, 1=Space, 2=Computers

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Convert text to features
vectorizer = CountVectorizer()
X_train_counts = vectorizer.fit_transform(X_train)
X_test_counts = vectorizer.transform(X_test)

# Train classifier
clf = MultinomialNB()
clf.fit(X_train_counts, y_train)

# Predict and evaluate
y_pred = clf.predict(X_test_counts)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

# 27. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_classification


X, y = make_classification(
    n_samples=300,
    n_features=2,
    n_redundant=0,
    n_informative=2,
    n_clusters_per_class=1,
    random_state=42
)

# 2. Different C values to test
C_values = [0.1, 1, 10, 100]

# 3. Plot decision boundaries
plt.figure(figsize=(12, 10))

for i, C in enumerate(C_values, 1):
    svm = SVC(kernel='linear', C=C)
    svm.fit(X, y)
    
    # For decision boundary
    ax = plt.subplot(2, 2, i)
    
    # Plot points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
    
    # Create grid
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))
    
    # Decision function values
    Z = svm.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Decision boundary & margins
    plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='k')
    plt.contour(xx, yy, Z, levels=[-1, 1], linewidths=1, colors='grey', linestyles='dashed')
    
    plt.title(f"SVM Decision Boundary (C = {C})")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")

plt.tight_layout()
plt.show()
```

# 28. Write a Python program to train a Bernoulli Naïve Bayes classifier for binary classification on a dataset with binary features

# 29. Write a Python program to apply feature scaling before training an SVM model and compare results with unscaled data


```python

import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score


df = sns.load_dataset('iris')

X = df.iloc[:, :-1]
y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)


model_unscaled = SVC(kernel='rbf')
model_unscaled.fit(X_train, y_train)
y_pred_unscaled = model_unscaled.predict(X_test)

acc_unscaled = accuracy_score(y_test, y_pred_unscaled)
print("Accuracy WITHOUT scaling:", acc_unscaled)



scaler = StandardScaler()

# Fit only on training data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model_scaled = SVC(kernel='rbf')
model_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = model_scaled.predict(X_test_scaled)

acc_scaled = accuracy_score(y_test, y_pred_scaled)
print("Accuracy WITH scaling:", acc_scaled)

print("\nComparison:")
print("Unscaled Accuracy:", acc_unscaled)
print("Scaled Accuracy:", acc_scaled)

if acc_scaled > acc_unscaled:
    print(" Scaling improved model performance")
elif acc_scaled == acc_unscaled:
    print("Scaling did not change the performance")
else:
    print(" Scaling decreased performance")
```

# 31. Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C, gamma, kernel)

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC

X, y = datasets.load_iris(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

params = {
    "C": [0.1, 1, 10],
    "gamma": ["scale", "auto", 0.1, 1],
    "kernel": ["linear", "rbf", "poly"]
}

grid = GridSearchCV(SVC(), params, cv=5)
grid.fit(X_train, y_train)

print(grid.best_params_)
print(grid.best_score_)
print(grid.score(X_test, y_test))
```

# 32. Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and check it improve accuracy

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

X, y = make_classification(n_samples=1000, n_features=10, weights=[0.9,0.1], random_state=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

model1 = SVC()
model1.fit(X_train, y_train)
pred1 = model1.predict(X_test)
acc1 = accuracy_score(y_test, pred1)

model2 = SVC(class_weight="balanced")
model2.fit(X_train, y_train)
pred2 = model2.predict(X_test)
acc2 = accuracy_score(y_test, pred2)

print("Accuracy without class weights:", acc1)
print("Accuracy with class weights:", acc2)

```

# 35. Write a Python program to perform feature selection before training a Naïve Bayes classifier and compare results

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

model_full = GaussianNB()
model_full.fit(X_train, y_train)
pred_full = model_full.predict(X_test)
acc_full = accuracy_score(y_test, pred_full)

selector = SelectKBest(score_func=f_classif, k=2)
X_new = selector.fit_transform(X, y)
X_train_sel, X_test_sel, y_train_sel, y_test_sel = train_test_split(X_new, y, test_size=0.3, random_state=1)

model_sel = GaussianNB()
model_sel.fit(X_train_sel, y_train_sel)
pred_sel = model_sel.predict(X_test_sel)
acc_sel = accuracy_score(y_test_sel, pred_sel)

print("Accuracy with all features:", acc_full)
print("Accuracy with selected features:", acc_sel)
```

# 37.  Write a Python program to train an SVM Classifier using Linear, Polynomial, and RBF kernels on the Breast Cancer dataset and compare their accuracy

# 40.Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and compare accuracy

```python


from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score

X, y = datasets.load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

svm_full = SVC(kernel='linear')
svm_full.fit(X_train, y_train)
pred_full = svm_full.predict(X_test)
acc_full = accuracy_score(y_test, pred_full)

selector = RFE(SVC(kernel='linear'), n_features_to_select=2)
X_train_rfe = selector.fit_transform(X_train, y_train)
X_test_rfe = selector.transform(X_test)

svm_rfe = SVC(kernel='linear')
svm_rfe.fit(X_train_rfe, y_train)
pred_rfe = svm_rfe.predict(X_test_rfe)
acc_rfe = accuracy_score(y_test, pred_rfe)

print("Accuracy with all features:", acc_full)
print("Accuracy with RFE-selected features:", acc_rfe)
```

# 43.Write a Python program to train an SVM Classifier and visualize the Confusion Matrix using seaborn

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

X, y = datasets.load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

model = SVC()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

# 46.  Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt

X, y = datasets.load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

model = SVC(kernel='linear', probability=True)
model.fit(X_train, y_train)
y_prob = model.predict_proba(X_test)[:, 1]

precision, recall, _ = precision_recall_curve(y_test, y_prob)
avg_precision = average_precision_score(y_test, y_prob)

plt.plot(recall, precision, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall Curve (AP={avg_precision:.2f})')
plt.show()
```
